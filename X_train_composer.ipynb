```python
import pandas as pd

davidson = pd.read_csv('./team_adv_stats_avg/davidson.csv')
duke = pd.read_csv('./team_adv_stats_avg/duke.csv')
fairleigh_dickinson = pd.read_csv('./team_adv_stats_avg/fairleigh-dickinson.csv')
florida_am = pd.read_csv('./team_adv_stats_avg/florida-am.csv')
florida_state = pd.read_csv('./team_adv_stats_avg/florida-state.csv')
hofstra = pd.read_csv('./team_adv_stats_avg/hofstra.csv')
illinois = pd.read_csv('./team_adv_stats_avg/illinois.csv')
indiana = pd.read_csv('./team_adv_stats_avg/indiana.csv')
iowa = pd.read_csv('./team_adv_stats_avg/iowa.csv')
maryland = pd.read_csv('./team_adv_stats_avg/maryland.csv')
michigan = pd.read_csv('./team_adv_stats_avg/michigan.csv')
michigan_state = pd.read_csv('./team_adv_stats_avg/michigan-state.csv')
minnesota = pd.read_csv('./team_adv_stats_avg/minnesota.csv')
nebraska = pd.read_csv('./team_adv_stats_avg/nebraska.csv')
new_orleans = pd.read_csv('./team_adv_stats_avg/new-orleans.csv')
northwestern = pd.read_csv('./team_adv_stats_avg/northwestern.csv')
ohio_state = pd.read_csv('./team_adv_stats_avg/ohio-state.csv')
penn_state = pd.read_csv('./team_adv_stats_avg/penn-state.csv')
purdue = pd.read_csv('./team_adv_stats_avg/purdue.csv')
rutgers = pd.read_csv('./team_adv_stats_avg/rutgers.csv')
wisconsin = pd.read_csv('./team_adv_stats_avg/wisconsin.csv')

davidson = davidson.loc[:, ~davidson.columns.str.startswith('Unnamed')]
duke = duke.loc[:, ~duke.columns.str.startswith('Unnamed')]
fairleigh_dickinson = fairleigh_dickinson.loc[:, ~fairleigh_dickinson.columns.str.startswith('Unnamed')]
florida_am = florida_am.loc[:, ~florida_am.columns.str.startswith('Unnamed')]
florida_state = florida_state.loc[:, ~florida_state.columns.str.startswith('Unnamed')]
hofstra = hofstra.loc[:, ~hofstra.columns.str.startswith('Unnamed')]
illinois = illinois.loc[:, ~illinois.columns.str.startswith('Unnamed')]
indiana = indiana.loc[:, ~indiana.columns.str.startswith('Unnamed')]
iowa = iowa.loc[:, ~iowa.columns.str.startswith('Unnamed')]
maryland = maryland.loc[:, ~maryland.columns.str.startswith('Unnamed')]
michigan = michigan.loc[:, ~michigan.columns.str.startswith('Unnamed')]
michigan_state = michigan_state.loc[:, ~michigan_state.columns.str.startswith('Unnamed')]
minnesota = minnesota.loc[:, ~minnesota.columns.str.startswith('Unnamed')]
nebraska = nebraska.loc[:, ~nebraska.columns.str.startswith('Unnamed')]
new_orleans = new_orleans.loc[:, ~new_orleans.columns.str.startswith('Unnamed')]
northwestern = northwestern.loc[:, ~northwestern.columns.str.startswith('Unnamed')]
ohio_state = ohio_state.loc[:, ~ohio_state.columns.str.startswith('Unnamed')]
penn_state = penn_state.loc[:, ~penn_state.columns.str.startswith('Unnamed')]
purdue = purdue.loc[:, ~purdue.columns.str.startswith('Unnamed')]
rutgers = rutgers.loc[:, ~rutgers.columns.str.startswith('Unnamed')]
wisconsin = wisconsin.loc[:, ~wisconsin.columns.str.startswith('Unnamed')]
```

```python
purdue_train_df = pd.DataFrame()

purdue = purdue.iloc[5:]
```

```python
# Iterate through each row in the purdue DataFrame
for index, purdue_row in purdue.iterrows():
    # Get the date and opponent from the current row
    date = purdue_row['Date']
    opp = purdue_row['Opp']

    # Get the opponent's stats based on the opponent's name
    if opp == 'Davidson':
        opp_stats = davidson
    elif opp == 'Duke':
        opp_stats = duke
    elif opp == 'Fairleigh Dickinson':
        opp_stats = fairleigh_dickinson
    elif opp == 'Florida A&M':
        opp_stats = florida_am
    elif opp == 'Florida State':
        opp_stats = florida_state
    elif opp == 'Hofstra':
        opp_stats = hofstra
    elif opp == 'Illinois':
        opp_stats = illinois
    elif opp == 'Indiana':
        opp_stats = indiana
    elif opp == 'Iowa':
        opp_stats = iowa
    elif opp == 'Maryland':
        opp_stats = maryland
    elif opp == 'Michigan':
        opp_stats = michigan
    elif opp == 'Michigan State':
        opp_stats = michigan_state
    elif opp == 'Minnesota':
        opp_stats = minnesota
    elif opp == 'Nebraska':
        opp_stats = nebraska
    elif opp == 'New Orleans':
        opp_stats = new_orleans
    elif opp == 'Northwestern':
        opp_stats = northwestern
    elif opp == 'Ohio State':
        opp_stats = ohio_state
    elif opp == 'Penn State':
        opp_stats = penn_state
    elif opp == 'Rutgers':
        opp_stats = rutgers
    elif opp == 'Wisconsin':
        opp_stats = wisconsin
    else:
        print('Opponent not found')

    # Get the opponent's stats for the specified date
    opp_stats = opp_stats[opp_stats['Date'] == date]
    opp_stats = opp_stats.iloc[0]

    # Get Purdue's stats for the current date
    purdue_stats = purdue_row

    # Concatenate the stats into the same row for Purdue and the opponent
    combined_stats = pd.concat([purdue_stats, opp_stats], axis=0, keys=['Purdue', 'Opp'])

    # Append the combined row to the purdue_train_df
    purdue_train_df = purdue_train_df.append(combined_stats, ignore_index=True)

# Resetting the index of the resulting DataFrame
purdue_train_df.reset_index(drop=True, inplace=True)

purdue_train_df
```
