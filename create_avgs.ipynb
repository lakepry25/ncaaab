```python
import copy, math, os
import pandas as pd
import numpy as np
```

```python
folder_path = './team_stats/'

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    csv_path = './team_adv_stats_avg/' + filename.split('_')[0] + '.csv'

    df = pd.read_csv(file_path, encoding='latin-1')

    df.drop(['Opp.1', 'W/L'], axis=1, inplace=True)

    df = df.loc[:, ~df.columns.str.startswith('Unnamed')]

    column_rename_mapping = {'Date': 'Date', 'Opp': 'Opp', 'Tm': 'Tm_Score', 'Opp.1': 'Opp_Score', 'ORtg': 'ORtg', 'DRtg': 'DRtg', 'Pace': 'Pace', 'FTr': 'FTr', '3PAr': '3PAr',
       'TS%': 'TS%', 'TRB%': 'TRB%', 'AST%': 'AST%', 'STL%': 'STL%', 'BLK%': 'BLK%', 'eFG%': 'eFG%', 'TOV%': 'TOV%', 'ORB%': 'ORB%', 'FT/FGA': 'FT/FGA',
       'eFG%.1': 'Opp_eFG%.1', 'TOV%.1': 'Opp_TOV%', 'DRB%': 'DRB%', 'FT/FGA.1': 'Opp_FT/FGA'}
    
    df.rename(columns=column_rename_mapping, inplace=True)

    averages_df = df.copy()

    for i in range(5, len(df)):
        # Calculate the mean of the previous 5 rows
        averages_df.loc[i, df.columns[3:]] = df.loc[i-5:i-1, df.columns[1:]].mean()

    averages_df = averages_df.iloc[5:]
    averages_df.to_csv(csv_path, index=False)
```
